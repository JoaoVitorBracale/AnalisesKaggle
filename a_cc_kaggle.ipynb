Creation of categorization functions for better interpretation / understanding of the dataset

```python
#Based on the description and additional information passed from the available dataset

def d_sex(sex):
    if sex == 1:
        return 'Male'
    elif sex == 2:
        return 'Female'

def d_education(edu):
    if edu == 1:
        return 'Graduate school'
    elif edu == 2:
        return 'University'
    elif edu == 3:
        return 'High schoo'
    elif edu == 4:
        return 'Other'
    elif edu == 5 or edu == 6:
        return 'Unknown'
    
def d_marriage(marriage):
    if marriage == 1:
        return 'Married'
    elif marriage == 2:
        return 'Single'
    elif marriage == 3:
        return 'Other'
    
def d_payment(pay):
    if pay == 0:
        return 'No'
    elif pay == 1:
        return 'Yes'
    
def pays(p):
    if p == -1:
        return 'Pay duly'
    elif p == 0:
        return 'Unknown'
    elif p == 1:
        return 'Payment delay for one month'
    elif p == 2:
        return 'Payment delay for two month'
    elif p == 3:
        return 'Payment delay for three month'
    elif p == 4:
        return 'Payment delay for four month'
    elif p == 5:
        return 'Payment delay for five month'
    elif p == 6:
        return 'Payment delay for six month'
    elif p == 7:
        return 'Payment delay for seven month'
    elif p == 8:
        return 'Payment delay for eight month'
    elif p == 9:
        return 'Payment delay for nine month and above'
```

```python
#Applying functions to their respective columns
        
ds_cc['SEX'] = ds_cc['SEX'].apply(d_sex)

ds_cc['EDUCATION'] = ds_cc['EDUCATION'].apply(d_education)

ds_cc['MARRIAGE'] = ds_cc['MARRIAGE'].apply(d_marriage)

ds_cc['default.payment.next.month'] = ds_cc['default.payment.next.month'].apply(d_payment)

ds_cc['PAY_0'] = ds_cc['PAY_0'].apply(pays)
ds_cc['PAY_2'] = ds_cc['PAY_2'].apply(pays)
ds_cc['PAY_3'] = ds_cc['PAY_3'].apply(pays)
ds_cc['PAY_4'] = ds_cc['PAY_4'].apply(pays)
ds_cc['PAY_5'] = ds_cc['PAY_5'].apply(pays)
ds_cc['PAY_6'] = ds_cc['PAY_6'].apply(pays)
 
```
